```python
# 4. 주어진 값으로 DataFrame 객체를 생성하시오

data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(data, index=labels)
```

```python
# 5. 객체내부 정보를 출력하시오
df
```

```python
# 6. 객체 상위 3열까지 출력하시오
df.head(3)
```

```python
# 7. animal과 age 컬럼만 출력하시오
df.loc[:, 'animal':'age']
```

```python
#8. 객체의 3, 4, 8번열에 해당하는 animal과 age 값만 출력 
df.loc[['c', 'd', 'h'], 'animal':'age']
```

```python
#9. visits 칼럼에서 3초과하는 값 출력

df[df['visits'] > 2]

    

    
```

```python
# 10. age 에서 NaN 값 출력
df[df['age'] == 'NaN' ]
   
```

```python
# 11. age가 3살 미만 고양이값 출력
df[(df['age'] < 3) & (df['animal'] == 'cat')]
```

```python
# 12. age가 2살이상 4살 미만인 값 출력
df[(df['age'] < 4) & (df['age'] >= 2)] 
```

```python
# 13. f 행의 나이를 1.5살로 변경
df.loc['f', 'age'] = 1.5
df
```

```python
# 14. 객체에서 visit 의 합 출력
df['visits'].sum()
```

```python
# 15. 동물별로 나이의 평균 출력
df.groupby('animal')['age'].mean()
```

```python
# 16. k행을 추가하여 dog , 5.5세, 우선권없음(no), 방문회수 2회 열을 추가
df.loc['k'] = ['dog', 5.5, 2, 'no' ]
df
```

```python
# 17. 객체에 있는 동물의 종류의 수 출력
df['animal'].nunique()
```

```python
# 18. age 는 내림차순, visits 는 오름차순으로 정렬
df.sort_values('age', ascending=False) 
```

```python
df.sort_values('visits') 
```

```python
# 19. priority 의 yes를 True, no 를 False  로 맵핑 후 출력
map = {'yes': True, 'no': False}
df['priority'] = df['priority'].map(map)
df
```

```python
# 20. snake 를 python 으로 값을 변경
df['animal'] = df['animal'].replace('snake', 'python')
df
```

```python
# 21. 각각의 동물 유형과 방문 횟수에 대해, 평균나이를 찾으시오. 
# 즉,각 행은 animal 이고, 각 열은 visit 이며, 값은 평균연령
# (힌트, 피벗 테이블을 활용해야 함)
df = df.pivot_table(index='animal', columns='visits', values='age', aggfunc='mean')
df

```
