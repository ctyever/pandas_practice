```python
import pandas as pd
import random
import string
import numpy as np
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def score():
    return np.random.randint(0,100)
```

```python
df = pd.DataFrame([[score() for i in range(1,6)] for i in range(1,11)], index=[id() for i in range(1,11)], columns=['국어', '영어', '수학', '과학', '사회'])
```

```python
print(df)
```

```python
#기술이 추가되었습니다 기존 학생들에게 랜덤으로 추가하시오
```

```python
df.loc[:, '기술'] = pd.Series([score() for i in range(1,11)], index=df.index)  
df
```

```python
# 각 학생들의 점수의 총합을 표현하는 컬럼을 추가하세요
```

```python
df['총합'] = df.sum(axis=1)
df
```

```python
df.loc['총점'] = [df['국어'].sum(), df['영어'].sum(), df['수학'].sum(), df['과학'].sum(), df['사회'].sum(), df['기술'].sum(), df['총합'].sum()]
df
```
